# Loading features that are present but not loaded.

There are quite a few features in the BHSA that are not loaded by default.
If you need them, you can load them easily.

```python
from tf.app import use
```

```python
A = use("bhsa", hoist=globals())
```

Let's ask for the list of all features.

```python
A.isLoaded()
```

You can load extra features in the following way.
See the [cheatsheet under `TF.load()`](https://annotation.github.io/text-fabric/tf/cheatsheet.html#loading).

```python
TF.load("lex0", add=True)
```

From Text-Fabric 9.1.5 onwards, you can also use a simpler call.
See the [cheatsheet under `A.load()`](http://localhost:8080/tf/cheatsheet.html#initialisation-configuration-meta-data-and-linking).

Let's also load `rank_occ` and `root`:

```python
A.load("rank_occ root")
```

We can now make use of the feature `root`:

```python
F.root.freqList()[0:10]
```
