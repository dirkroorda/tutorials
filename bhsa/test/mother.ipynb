First determine the from-types and the to-types of nodes connected by a mother edge:

A query by Oliver Glanz.

In MQL on version 4b:

```
select all objects where
[sentence
  [clause
    [phrase FOCUS
      [phrase as motherPhrase
        [word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
      ]
    ]
  ]
  ..
  [clause FOCUS rela = Attr AND mother = motherPhrase.self]
]
```

````python
queryPlain = """
sentence
  cl1:clause
    p1M:phrase
      word lex=JHWH/|>LHJM/|>L/
  cl2D:clause rela=Attr
"""
````

````python
resultsPlain = A.search(queryPlain)
````

````python
resultsPlain4b = A4b.search(queryPlain)
````

````python
query = """
sentence
  cl1:clause
    p1M:phrase
      word lex=JHWH/|>LHJM/|>L/
  cl2D:clause rela=Attr
p1M <mother- cl2D
"""
````

````python
results = A.search(query)
````

````python
results4b = A4b.search(query)
````

````python
A.show(results, start=1, end=2, condensed=True)
````
